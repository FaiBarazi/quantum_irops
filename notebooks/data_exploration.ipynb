```python
import polars as pl
```

```python
# data files: Note this is not commited to the repo yet. Need to check if it is ok to add the data. 
canceled_flights = "./data/PRMI-DM_TARGET_FLIGHTS.csv"
available_flights = "./data/PRMI-DM-AVAILABLE_FLIGHTS.csv"
all_pnrs = "./data/PRMI_DM_ALL_PNRs.csv"

df_cancelled = pl.read_csv(canceled_flights)
df_all_flights = pl.read_csv(available_flights)
df_pnrs = pl.read_csv(all_pnrs)
```

```python
# Number of business class and economy. "Y": stands for economy , "C": Business
df_pnrs.group_by("CABIN_CD").len()
```

```python
df_pnr_cancelled = df_pnrs.join(df_cancelled, on="DEP_KEY", how="semi")
total_pass_affected = df_pnr_cancelled["PAX_CNT"].sum()
print("Number of cancelled PNRs: ", len(df_pnr_cancelled))
print("Number of passengers affected: ", total_pass_affected)
```

```python
# Join on cancelled flight keys - left join on all flights. Anti meaning inverse, all the keys that are NOT common. 
df_flight_available = df_all_flights.join(df_cancelled, on="DEP_KEY", how="anti")

# Sort df by Departure Key
df_flight_available = df_flight_available.sort(by="DEP_KEY")
# Filter based on flights with available seats. 
df_flight_available = df_flight_available.filter(pl.all_horizontal(pl.col("C_AVAIL_CNT", "Y_AVAIL_CNT") > 0))

# Count Available flights with seats.
print("Available flights: ", len(df_flight_available))

# Count avialable seats for business and economy
available_business = df_flight_available["C_AVAIL_CNT"].sum()
available_economy = df_flight_available["Y_AVAIL_CNT"].sum()
print(
    f"""
Available Seats: 
- Economy: {available_economy}
- Business: {available_business}
"""
)
```

Q is the Hamiltonian. Described below: 
let i be the PNR index and j the flight index.

$\lambda_1\sum_{f}\sum{_{p}}T_{pf}Dst_{pf}Dep_{pf}x_{pf}+ \lambda_2\sum{_{p}}(\sum_{f}x_{pf}-1)^2 + \sum_f(-\lambda_3(S_f - \sum_p s_px_{i,j})+\lambda_4(S_f - \sum_p s_px_{i,j})^2)$


The diagnoal linear interaction. 

Where: 
- $\lambda$ is cost.
- $T_px$: whatever equation we use for time.
- $Dst_{px}$: Boolean- destination matches PNR
- $Dep_{px}$: Boolean - departure mathces PNR
- $S_f$: seats available on flight.
