```python
# data files: Note this is not commited to the repo yet. Need to check if it is ok to add the data. 
canceled_flights = "./data/PRMI-DM_TARGET_FLIGHTS.csv"
available_flights = "./data/PRMI-DM-AVAILABLE_FLIGHTS.csv"
all_pnrs = "./data/PRMI_DM_ALL_PNRs.csv"

df_cancelled = pl.read_csv(canceled_flights)
df_available = pl.read_csv(available_flights)
df_pnrs = pl.read_csv(all_pnrs)
```

```python
# Number of business class and economy. "Y": stands for economy , "C": Business
df_pnrs.group_by("CABIN_CD").len()
```

```python
# Counting aviable seats in available flights: 
available_c_seats = df_available.select(
    pl.col("C_AVAIL_CNT").filter(pl.col("C_AVAIL_CNT") > 0).sum()
).item()

available_y_seats = df_available.select(
    pl.col("Y_AVAIL_CNT").filter(pl.col("Y_AVAIL_CNT") > 0).sum()
).item()
print(
    f"""
Available: 
- Business: {available_c_seats}
- Economy: {available_y_seats}
- Total Avialable : {available_c_seats + available_y_seats}

""")
```
